# The Series Data Structure

```python
import pandas as pd

#Creating series

students = ['Alice','Jack','Molly']

print(pd.Series(students))

#pandas stores series values in a typed array using numpy

numbers = [1,2,3]

print(pd.Series(numbers))

#treating None values in series

students = ['Alice','Jack',None]

print(pd.Series(students))

numbers = [1,2,None]

#the the type of this series is now float64 instead int64
print(pd.Series(numbers))
#Pandas represents NaN as a floating point number, and the integers are
#typecast to floats
#But None is different of NaN for Python

import numpy as np 

print(np.nan == None)
print(np.nan == np.nan) #this is crazy
#you need to use special functions to test dor the presenc of a NaN
np.isnan(np.nan)

#creatinf labeled series
students_scores = {'Alice':'Physics','Jack':'Chemistry','Molly':'English'}
s = pd.Series(students_scores)
s.index

#dtype of objct is not just for strings but for arbitrary objects

s = pd.Series(['Physics','Chemistry','English'], index = ['Alice','Jack','Molly'])
print(s)



```

# Querying a Series

```python
#a pandas series can be queried either by the index position 
#or the index label. If you don't give an index to the series 
#when querying the position and label are effctively the same values
#querying by postion: iloc attribute (are ATTRIBUTES not methods)
#querying by label: loc attribute


students_scores = {'Alice':'Physics','Jack':'Chemistry','Molly':'English'}

s = pd.Series(students_scores)

print(s.iloc[2])
print(s.loc['Molly'])
print(s[2])
print(s['Molly'])

#it is better use the iloc and loc attributes

class_code = {99:'Physics',100:'Chemistry'}

s = pd.Series(class_code)

#print(s[0]) #error: pandas try to use s.loc[0] because 


#working with the data with pandas
grades = pd.Series([90,80,70,80])

total = np.sum(grades)
print(total/len(grades))


numbers = pd.Series(np.random.randint(0,1000,10000))

numbers+=2
#vectorizagtion is the ability for a computer to execute 
#multiple instructions at once and with high performance chips 
#specially graphics cards, you can get dramatic speedups

#the .loc attribute lets you not only modify data in place 
#but also add new data as well

s = pd.Series([1,2,3])

s.loc['History'] = 102

print(s)

#index in Series can have the same indexing

students_classes = pd.Series(students_scores)

kelly_classes = pd.Series(['Philosophy','Arts','Math'], index = ['Kelly','Kelly','Kelly'])

print(kelly_classes)

all_students_classes = students_classes.append(kelly_classes)

print(all_students_classes)
print(all_students_classes['Kelly'])

#but the students_classes wasn't changed

print(students_classes)
```

```python
import matplotlib.pyplot as plt
import mplleaflet
import pandas as pd
from IPython.display import display

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

def leaflet_plot_stations(binsize, hashid):

    df = pd.read_csv('data/C2A2_data/BinSize_d{}.csv'.format(binsize))
#     display(df.head())
    station_locations_by_hash = df[df['hash'] == hashid]

    lons = station_locations_by_hash['LONGITUDE'].tolist()
    lats = station_locations_by_hash['LATITUDE'].tolist()

    plt.figure(figsize=(8,8))

    plt.scatter(lons, lats, c='r', alpha=0.7, s=200)

    return mplleaflet.display()

leaflet_plot_stations(400,'fb441e62df2d58994928907a91895ec62c2c42e6cd075c2700843b89')
```

```python
import numpy as np
from datetime import datetime as dt
# from datetime import date, timedelta

binsize = 400
hashid = 'fb441e62df2d58994928907a91895ec62c2c42e6cd075c2700843b89'
df = pd.read_csv('data/C2A2_data/BinnedCsvs_d{}/{}.csv'.format(binsize,hashid))

pattern = '\d\d\d\d-(\d\d-\d\d)'
df['Month-Day'] = df['Date'].str.extract(pattern,expand=False)
# display(df.head())
df['Date'] = pd.to_datetime(df['Date'])
df['Year'] = df['Date'].dt.year
df = df[df['Month-Day']!='02-29']
    
df2 = df[df['Year']!=2015]
dfmax = df2.groupby(['Month-Day']).agg({"Data_Value":np.max})
dfmin = df2.groupby(['Month-Day']).agg({"Data_Value":np.min})

dfmin['Data_Value'] = dfmin['Data_Value']/10.0
dfmax['Data_Value'] = dfmax['Data_Value']/10.0


df3 = df[df['Year']==2015]
dfmax2015 = df3.groupby(['Month-Day']).agg({"Data_Value":np.max})
dfmin2015 = df3.groupby(['Month-Day']).agg({"Data_Value":np.min})
dfmin2015['Data_Value'] = dfmin2015['Data_Value']/10.0
dfmax2015['Data_Value'] = dfmax2015['Data_Value']/10.0

dfmin2015.reset_index(inplace=True)
dfmin2015.reset_index(inplace=True)
dfmin2015.set_index('Month-Day',inplace=True)
dfmin.reset_index(inplace=True)
dfmin.reset_index(inplace=True)
dfmin.set_index('Month-Day',inplace=True)

min2015 = dfmin2015[dfmin2015['Data_Value']<dfmin['Data_Value']]

dfmax2015.reset_index(inplace=True)
dfmax2015.reset_index(inplace=True)
dfmax2015.set_index('Month-Day',inplace=True)
dfmax.reset_index(inplace=True)
dfmax.reset_index(inplace=True)
dfmax.set_index('Month-Day',inplace=True)

max2015 = dfmax2015[dfmax2015['Data_Value']>dfmax['Data_Value']]
```

```python
from itertools import accumulate

plt.figure()

# print(dfmin.Data_Value)
plt.plot(np.arange(365),dfmin.Data_Value,linestyle='--', color='blue',alpha=0.5, label='Record Low')
plt.plot(np.arange(365),dfmax.Data_Value,linestyle='--', color='red',alpha=0.5, label='Record High')
plt.gca().fill_between(range(365),
                       dfmin.Data_Value, dfmax.Data_Value, 
                       facecolor='blue',alpha=0.15)

xlabels = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
nDaysOfMonth = [31,28,31,30,31,30,31,31,30,31,30,31]
xvals = [0]+list(accumulate(nDaysOfMonth[:-1]))
xvals = np.array(xvals)
plt.xticks(xvals,xlabels,alpha=0.8)
# plt.tick_params(bottom = False)


plt.xlim([0,364])
plt.ylim([-42,42])
ax = plt.gca()
# ax.set_xlabel('Da')
# ax.set_ylabel('Temperature (°C)')
ax.set_title('Daily record high and low temperatures (°C)\n over 10 years (2005-2014)')


plt.scatter(min2015['index'],min2015['Data_Value'], s=10, c='blue',label='2015 New Record Low')
plt.scatter(max2015['index'],max2015['Data_Value'], s=10, c='red',label='2015 new Record High')

plt.legend()

# plt.savefig('plotAss2')

plt.show()
```

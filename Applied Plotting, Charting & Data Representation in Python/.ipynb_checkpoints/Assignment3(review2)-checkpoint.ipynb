```python
# Use the following data for this assignment:

# Backend
%matplotlib ipympl

# Load packages
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import pandas as pd
import numpy as np

np.random.seed(12345)

df = pd.DataFrame([np.random.normal(32000,200000,3650), 
                   np.random.normal(43000,100000,3650), 
                   np.random.normal(43500,140000,3650), 
                   np.random.normal(48000,70000,3650)], 
                  index=[1992,1993,1994,1995])
df = df.transpose()
df.head()
```

```python
# Sample size
n = df.shape[0]

# Calculate the average values
mean_values = df.mean()

# Calculate the standard deviations
alpha = 0.05
std_values = df.std()

# Error bars according TLC
error = std_values/np.sqrt(n) 

# Create lists for the plot
years = list(df.columns)
x_pos = np.arange(len(years))

# Colors
my_colors = plt.cm.get_cmap('RdBu_r')
my_norm = mcolors.Normalize(vmin=-mean_values.max(), vmax=mean_values.max())

# Colors function for bars
def color_bars(val, rectangles, cmap, norm):
    heights = np.array([i.get_height() for i in rectangles])
    diff = heights - val
    colors = cmap(norm(diff))
    for rectangle, color in zip(rectangles, colors):
        rectangle.set_facecolor(color)
    fig.canvas.draw()

# Build the plot
fig, ax = plt.subplots()
# Bars
my_bar = ax.bar(x_pos, mean_values, yerr=error, align='center', alpha=0.5, edgecolor='black', capsize=15)
hline = ax.axhline(y=0, c='lightslategray', linestyle=':')
ax.set_ylabel('Value of distribution')
ax.set_xticks(x_pos)
ax.set_xticklabels(years)
ax.set_title('Generated data between 1992-1995')
ax.yaxis.grid(False)

# remove the frame of the chart
for spine in plt.gca().spines.values():
    spine.set_visible(False)

# Interaction
def onclick(event):
    if event.inaxes:
        fig.suptitle('{:.2f}'.format(event.ydata))
        hline.set_ydata([event.ydata, event.ydata])
        color_bars(event.ydata, my_bar, cmap=my_colors, norm=my_norm)
        fig.canvas.draw()

fig.canvas.mpl_connect('button_press_event', onclick)

plt.show()
```

```python
# Save image in png format
fig.savefig('Interactivity.png')
```

```python
%matplotlib notebook
import networkx as nx
import matplotlib.pyplot as plt

G=nx.Graph()

G.add_edge('F','B')
G.add_edge('F','A')
G.add_edge('F','C')
G.add_edge('F','D')
G.add_edge('F','E')
G.add_edge('A','B')
G.add_edge('A','C')
G.add_edge('A','D')
G.add_edge('B','C')
G.add_edge('C','D')
G.add_edge('D','E')
G.add_edge('A','E')

plt.figure(figsize=(7,8))
pos = nx.circular_layout(G)
nx.draw_networkx(G, pos)

```

```python
nx.clustering(G, 'F')
```

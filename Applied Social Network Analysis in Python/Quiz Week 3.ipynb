```python
%matplotlib notebook
import networkx as nx
import matplotlib.pyplot as plt


G=nx.DiGraph()

G.add_edge('A','B')
G.add_edge('B','A')
G.add_edge('A','C')
G.add_edge('C','D')
G.add_edge('D','C')

plt.figure(figsize=(7,8))
# pos = nx.circular_layout(G)
nx.draw_networkx(G)
```

```python
print(nx.pagerank(G, alpha=0.5))
print(nx.pagerank(G, alpha=0.8))
print(nx.pagerank(G, alpha=0.9))
print(nx.pagerank(G, alpha=0.95))
```
